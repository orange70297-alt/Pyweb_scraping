```python
# selenium 起手式
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
# from webdriver_manager.chrome import ChromeDriverManager

# 建立 Service 物件，指定 chromedriver.exe 的路徑
service = Service()

# 設定 Chrome 瀏覽器的選項
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized") # Chrome 瀏覽器在啟動時最大化視窗
options.add_argument("--incognito") # 無痕模式
options.add_argument("--disable-popup-blocking") # 停用 Chrome 的彈窗阻擋功能。

# 建立 Chrome 瀏覽器物件
driver = webdriver.Chrome(options=options)
driver.get("https://www.momoshop.com.tw/main/Main.jsp")

import time
```

```python
driver.find_element(By.CSS_SELECTOR, "input[name='search-input']").send_keys("潛水蛙鞋")
driver.find_element(By.CSS_SELECTOR, ".flex.h-full.items-center.overflow-hidden > button").click()

```

```python
#計算整個頁面的數量 
item_list = driver.find_elements(By.CLASS_NAME, "listAreaLi")
print(item_list)
print(len(item_list))
```

```python
#抓到第一筆的詳細資料

Data = []
Product_Title = driver.find_element(By.CSS_SELECTOR, ".listAreaLi h3").text
Product_Price = driver.find_element(By.CSS_SELECTOR, ".prdInfoWrap b").text
Product_Img = driver.find_element(By.CSS_SELECTOR, ".prdImg").get_attribute("src")
Product_href = driver.find_element(By.CSS_SELECTOR, ".prdImgWrap a").get_attribute("href")



print(Product_Title)
print(Product_Price)
print(Product_Img)
print(Product_href)
```
