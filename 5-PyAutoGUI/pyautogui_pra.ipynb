```python
import keyboard
import pyautogui
import time

"""
取得起始點的方式
1. time.sleep(2) 搭配 pyautogui.position() 取得滑鼠位置
2. 在 cmd 使用 python -m pyautogui 進入互動模式
"""

# 設定起始點
start_point = (1100,380)
# 移動滑鼠到起始點
pyautogui.moveTo(start_point[0], start_point[1])

distance = 200
# 按下滑鼠左鍵
while distance > 0:
# 向右拖曳
    pyautogui.dragRel(distance,0)
    # 向下拖曳
    distance -= 10
    pyautogui.dragRel(0,distance)
    # 向左拖曳
    pyautogui.dragRel(-distance,0)
    # 向上拖曳
    distance -= 10
    pyautogui.dragRel(0,-distance)
# 放開滑鼠左鍵
```

### 螢幕截圖 與 圖片辨識

```python
import pyautogui
# 擷取指定區域螢幕截圖                               (x, y, width, height)
pyautogui.screenshot('test_screenshot.png', region=(1088, 353, 300, 250))  

# 在螢幕上尋找圖片並點擊

location = pyautogui.locateCenterOnScreen('test_screenshot.png', confidence=0.8 )

if location :
    pyautogui.click(location)
```

```python
import pyautogui
"""
小算盤
擷取數字圖片後自動點擊
"""

# 要擷取數字的大小
num_size = 30
# 數字之間的水平與垂直距離
x_distance = 80
y_distance = 50

# 起始位置 (從數字1的位置開始)
start_location = (30, 410)
for i in range(3):
     pyautogui.screenshot(f"{1+i}.png" , region=(start_location[0]+ x_distance*i,start_location[1],num_size,num_size))
     pyautogui.screenshot(f"{4+i}.png" , region=(start_location[0]+ x_distance*i,start_location[1]-y_distance,num_size,num_size))
     pyautogui.screenshot(f"{7+i}.png" , region=(start_location[0]+ x_distance*i,start_location[1]-y_distance*2,num_size,num_size))
    
pyautogui.screenshot("0.png" , region=(100 , 465 , 80, 50))
        
```

```python
# 執行點擊動作
def click_num(num):
    location = pyautogui.locateCenterOnScreen(f"{num}.png")
    if location:
        pyautogui.click(location)
 
click_num(5)
click_num(2)
click_num(0)
```
