```python
import requests
from bs4 import BeautifulSoup as bs 
url = 'https://www.ptt.cc/bbs/Gossiping/index.html'

prefix = "https://www.ptt.cc"

r = requests.get(url)
soup = bs(r.text, 'lxml')

#containers = soup.select(".r-ent")
#containers = soup.select(".r-ent .title")
#print(container[0])

title_elements = soup.select(".r-ent .title a") #空一格往內找一層 (CSS) 

# print(title_elements[0])

for ele in title_elements: 
    print(f"標題:{ele.get_text()} 網址:{prefix + ele['href']}")

    ppt_response = requests.get(prefix + ele['href'])
    print(ppt_response.text)
    break
    

```
