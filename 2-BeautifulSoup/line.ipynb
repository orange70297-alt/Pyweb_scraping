```python
import json

import os 
folder_name = "line_stickers"
print(os.path.exists(folder_name))

#建立資料夾
if not os.path.exists(folder_name):
    os.mkdir("line_stickers")

    url = 'https://store.line.me/stickershop/product/29386925/zh-Hant'

def get_stickers(url):
    
    r = requests.get(url)
    soup = bs(r.text, 'lxml')

    # 找到貼圖清單
    sticker_list = soup.find("ul", class_="mdCMN09Ul FnStickerList")
    # 找到貼圖元素 
    sticker_elements = sticker_list.find_all("li")
    print(len(sticker_elements))

    for index, sticker_element in enumerate(sticker_elements, start=1):

        data_preview = json.loads(sticker_element["data-preview"])
        static_url = data_preview["staticUrl"]
        
        image_content = requests.get(static_url).content
        with open (f"line_stickers/line_sticker{index}.png", mode="wb")as f:
            f.write(image_content)
```
