```python
import requests
from bs4 import BeautifulSoup as bs 

def get_stock_info(code):
    url = f"https://tw.stock.yahoo.com/quote/{code}.TW"
    r = requests.get(url)
    soup = bs(r.text, 'lxml')

    up_or_down_mapping = {
        "border-color:transparent transparent #ff333a transparent;border-width:0 6.5px 9px 6.5px": "漲",
        "border-color:#00ab5e transparent transparent transparent;border-width:9px 6.5px 0 6.5px": "跌" }
    
        
    
    Title = soup.find("h1", class_="C($c-link-text) Fw(b) Fz(24px) Mend(8px) Whs(nw)")

    title = Title.text

    price = soup.find('span', class_="Fz(32px)").text
   
    up_down_style = soup.find('span', class_="Mend(4px) Bds(s)")['style']

    percent = soup.find("span", class_="Jc(fe)").text
    
    up_or_down = up_or_down_mapping.get(up_down_style)

    print(f"{title} {price} {up_or_down} {percent}")


```

```python
stock_codes = ["2317","2313"]
for stock_code in stock_codes:
    get_stock_info(stock_code)
```
