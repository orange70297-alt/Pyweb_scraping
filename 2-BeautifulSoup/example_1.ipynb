```python
import requests
from bs4 import BeautifulSoup as bs 

url = "https://tw.stock.yahoo.com/quote/2330.TW"
r = requests.get(url)
soup = bs(r.text, 'lxml')

up_or_down_mapping = {
    "border-color:transparent transparent #ff333a transparent;border-width:0 6.5px 9px 6.5px": "漲",
    "border-color:#00ab5e transparent transparent transparent;border-width:9px 6.5px 0 6.5px": "跌" }
```

```python
Title = soup.find("h1", class_="C($c-link-text) Fw(b) Fz(24px) Mend(8px) Whs(nw)").text
print(Title)

price = soup.find('span', class_="Fz(32px) Fw(b) Lh(1) Mend(16px) D(f) Ai(c) C($c-trend-down)").text
print(price)

up_down_style = soup.find('span', class_="Mend(4px) Bds(s)")['style']
print(up_down_style)

percent = soup.find(class_="Jc(fe) Fz(20px) Lh(1.2) Fw(b) D(f) Ai(c) C($c-trend-down)").text
print(percent)
up_or_down = up_or_down_mapping.get(up_down_style)

print(f"{Title} {price} {up_or_down} {percent}")



```
