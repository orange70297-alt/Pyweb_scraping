```python
# 引入 requests 模組
import requests

# 測試用 API 網址
url = 'https://httpbin.org/'

# 發送 GET 請求並取得回應
response = requests.get(url)

if response.status_code !=200:
    print('回傳異常')

# 回應的狀態碼
print(response)

# # 回應的原始內容（位元組）
print(response.status_code)

# # 回應的文字內容
print(response.text)
# print(response.headers)
```

```python
# 使用 GET 搭配參數
#' https://httpbin.org/get '
r = requests.get(url+'get', params={'city': "Taipei"})
print(r.text)

print(r.json())
data = r.json()
data['args']


# 等同於使用下面的寫法
r = requests.get(url+'get?city=Taipei')

print(r.json())
```

```python
# 使用 POST 搭配 data (常用在表單提交)
r = requests.post(url+'post',data={'username':'Jarvis'})
print(r.json())


```

```python
# 使用 POST 搭配 JSON (常用在 API 資料傳輸)
r = requests.post(url+'post' , json{})

r = requests.post(url+'post', json={'username': 'Jarvis'})
print(r.json())

```

```python
import requests
import json
"""
課堂練習 
參考網址: 'https://www.einvoice.nat.gov.tw/portal/ods/ODS318E/einvoice_open_data#/消費通路發票統計/ConsumpInvoiceStatic'
"""

base_url = 'https://dataset.einvoice.nat.gov.tw/ods/portal'
resource = 'api/v1/ConsumpInvoiceStatic'

r = requests.get(f'{base_url}/{resource}', params={'hsnNm': '新北市'})
print(r.json())

print(json.dumps(r.json(), indent=4, ensure_ascii=False))
```
